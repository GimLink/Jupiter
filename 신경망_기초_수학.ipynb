```python
import math
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
```

```python
def linear_function(x):
  a = 0.5
  b = 2

  return a * x + b
```

```python
print(linear_function(5))
```

```python
x = np.arange(-5,5,0.1)
y = linear_function(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Function')
```

```python
# y = ax2 + bx + c

def quadratic_function(x):
  a = 1
  b = -1
  c = -2

  return a*x**2 + b*x + c
```

```python
print(quadratic_function(2))
```

```python
y = quadratic_function(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Quadratic Function')
```

```python
# 3차 함수*다항함수) y = ax3 + bx2 +cx +d
def cubic_function(x):
  a = 4
  b = 0
  c = -1
  d = -8

  return a*x**3 + b*x**2 + c*x + d
```

```python
y = cubic_function(x)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Cubic Function')
```

```python
def my_func(x):
  a = 1
  b = -3
  c = 10

  return a*x**2 + b*x + c # 2차 함수 수식
```

```python
x = np.arange(-10 , 10 , 0.1)

y = my_func(x)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('My Function')
```
